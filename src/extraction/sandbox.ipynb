```python
import psycopg2
from lock_parameters import username_pg, password_pg, hostname_pg, port_pg, database_pg

connection_string = {
    "host": hostname_pg,
    "port": port_pg,
    "database": database_pg,
    "user": username_pg,
    "password": password_pg,
}

conn = psycopg2.connect(**connection_string)
cursor = conn.cursor()

# 1. Combien de documents contiennent "transbronch" dans le texte ?
print("🔍 Recherche directe dans la base...")

cursor.execute("""
    SELECT COUNT(*) 
    FROM dwh.dwh_document 
    WHERE UPPER(displayed_text) LIKE '%%TRANSBRONCH%%'
""")
print(f"  TRANSBRONCH dans displayed_text : {cursor.fetchone()[0]}")

cursor.execute("""
    SELECT COUNT(*) 
    FROM dwh.dwh_document 
    WHERE UPPER(displayed_text) LIKE '%%TRANSBRONCH%%'
    AND document_type = 'EXT'
""")
print(f"  + filtre document_type='EXT' : {cursor.fetchone()[0]}")

cursor.execute("""
    SELECT COUNT(*) 
    FROM dwh.dwh_document 
    WHERE UPPER(displayed_text) LIKE '%%TRANSBRONCH%%'
    AND document_type = 'EXT'
    AND title LIKE '%%Anapath%%'
""")
print(f"  + filtre title LIKE Anapath : {cursor.fetchone()[0]}")

# 2. Quels titres contiennent des BTB ?
cursor.execute("""
    SELECT title, COUNT(*) as cnt
    FROM dwh.dwh_document 
    WHERE UPPER(displayed_text) LIKE '%%TRANSBRONCH%%'
    GROUP BY title
    ORDER BY cnt DESC
    LIMIT 20
""")
print(f"\n📋 Titres des documents contenant 'transbronch' :")
for row in cursor.fetchall():
    print(f"  {row[1]:5d} | {row[0]}")

# 3. Documents avec texte NULL ?
cursor.execute("""
    SELECT COUNT(*) 
    FROM dwh.dwh_document 
    WHERE document_type = 'EXT'
    AND title LIKE '%%Anapath%%'
    AND displayed_text IS NULL
""")
print(f"\n⚠️  Documents Anapath avec displayed_text NULL : {cursor.fetchone()[0]}")

cursor.close()
conn.close()
```
