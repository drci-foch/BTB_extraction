```python
# Assuming you already have your DataFrames loaded as df_landscape and df_LBA
df_LBA = pd.read_excel(r"C:\Users\benysar\Documents\GitHub\BTB_extraction\src\output\landscape_structuration.xlsx")
df_landscape = pd.read_excel(r"C:\Users\benysar\Documents\GitHub\BTB_extraction\TBB_LANDSCAPE_whole.xlsx")

# Convert both columns to string type for consistent comparison
df_landscape['N sans P'] = df_landscape['N sans P'].astype(str)
df_LBA['Biopsy ID'] = df_LBA['Biopsy ID'].astype(str)

# Now perform the matching
matches = df_landscape[df_landscape['N sans P'].isin(df_LBA['Biopsy ID'])]

# Create a new DataFrame with only the matching rows
result = pd.merge(
    df_landscape['N sans P'], 
    df_LBA[['Biopsy ID', 'Prescripteur']],
    left_on='N sans P',
    right_on='Biopsy ID',
    how='inner'
)

# Print the number of matches
print(f"Number of matches found: {len(result)}")

# Print the matching values
print("\nMatching values:")
print(result)

# To see non-matches:
non_matches_landscape = df_landscape[~df_landscape['N sans P'].isin(df_LBA['Biopsy ID'])]
non_matches_LBA = df_LBA[~df_LBA['Biopsy ID'].isin(df_landscape['N sans P'])]

print("\nValues in 'N sans P' that don't match:")
print(non_matches_landscape['N sans P'])

print("\nValues in 'Biopsy ID' that don't match:")
print(non_matches_LBA['Biopsy ID'])
```

```python
import pandas as pd
from datetime import timedelta
import glob

file_pattern = r"C:\Users\benysar\Documents\GitHub\BTB_extraction\src\output\concatenate_all_landscape_review/*.xlsx"
files = glob.glob(file_pattern)

# Read and combine all files
dataframes = []
for file in files:
    df_temp = pd.read_excel(file)
    df_temp['source_file'] = file  # Add source tracking
    dataframes.append(df_temp)

# Combine into single dataframe
df = pd.concat(dataframes, ignore_index=True)
```
